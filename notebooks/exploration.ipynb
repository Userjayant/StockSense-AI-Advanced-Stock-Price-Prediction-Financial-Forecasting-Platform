# 📊 Stock Data Exploration
Quick EDA notebook for any ticker.

```python
import sys; sys.path.insert(0,'../src')
from data_loader import fetch_stock_data
from features import add_technical_indicators
import matplotlib.pyplot as plt

df = fetch_stock_data('AAPL', period='2y')
df.tail()
```

```python
feat = add_technical_indicators(df)
fig, ax = plt.subplots(figsize=(12,5))
ax.plot(feat.index, feat['Close'], label='Close')
ax.plot(feat.index, feat['SMA_20'], label='SMA 20')
ax.plot(feat.index, feat['SMA_50'], label='SMA 50')
ax.legend(); plt.title('AAPL Price + Moving Averages'); plt.show()
```
